```
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 11)
y = x ** 2
fig = plt.figure(figsize=(8,4), dpi=100)

fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.grid()
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```
