```python
#import sys
#sys.path.append("/anaconda3/lib/python3.6/site-packages") # where or which your anaconda path

import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('Online Retail.xlsx')
```

```python
df.head()
```

```python
# Clean up spaces in description and remove any rows that don't have a valid invoice
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
```

```python
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
basket = (df[df['Country'] =="France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
basket.head()
```

```python
# Show a subset of columns
basket.iloc[:,[0,1,2,3,4,5,6, 7]].head()
```

```python
# Convert the units to 1 hot encoded values
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1    
```

```python
# convert to one hot vector
basket_sets = basket.applymap(encode_units)
```

```python
# No need to track postage
# Remove column "Postage" (1 column)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
basket_sets.head()
```

```python
# Build up the frequent items
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
```

```python
frequent_itemsets.head()
```

```python
# Create the rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules[ (rules['lift'] >= 6) &
       (rules['confidence'] >= 0.8) ]
```

```python
basket2 = (df[df['Country'] =="Germany"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
# convert to one hot vector
basket_sets2 = basket2.applymap(encode_units)
```

```python
basket_sets2.drop('POSTAGE', inplace=True, axis=1)
```

```python
frequent_itemsets2 = apriori(basket_sets2, min_support=0.05, use_colnames=True)
```

```python
rules2 = association_rules(frequent_itemsets2, metric="lift", min_threshold=1)
rules2
```

```python
rules2[ (rules2['lift'] >= 4) &
        (rules2['confidence'] >= 0.5) ]
```
