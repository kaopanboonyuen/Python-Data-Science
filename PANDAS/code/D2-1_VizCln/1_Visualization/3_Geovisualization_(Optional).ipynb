```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_excel('https://github.com/davidjohnnn/all_datasets/raw/master/bay/Example_for_training-original.xlsx',sheetname='Data')
data = pd.DataFrame(data)
```

```python
data = data
data = data[data.PROVINCE != '(blank)']
data = data.reset_index(drop=True)
print(data.shape[0])
data.tail()
```

```python
lat_list = []
lon_list = []

for i in range( 0, data.shape[0]):
    lat_list.append(th_province_dict[data['PROVINCE'][i]][0])
    lon_list.append(th_province_dict[data['PROVINCE'][i]][1])

print(len(lat_list))
```

```python
data['latitude'] = lat_list
data['longitude'] = lon_list
data.tail()
```

```python
data.to_excel('Example_for_training_modify.xlsx',sheet_name='Data')
```

```python
data['ACTIVE_OS_NEW'] = data['ACTIVE_OS']*2

data['ACTIVE_OS_NEW_2'] = (data['ACTIVE_OS']-data['ACTIVE_OS'].min())/(data['ACTIVE_OS'].max()-data['ACTIVE_OS'].min())*100

data.tail()
```

```python
import matplotlib.pyplot as plt
data.plot(kind="scatter", x="longitude", y="latitude", alpha=0.6)
plt.show()
```

```python
# จำนวนบัญชีที่มียอดหนี้คงค้างและยังไม่ตัดหนี้สูญ
data.plot(kind="scatter", x="longitude", y="latitude",
    s=data['ACTIVE_ACCTS']*100, label="ACTIVE_ACCTS",
    c="ACTIVE_ACCTS", cmap=plt.get_cmap("jet"),
    colorbar=True, alpha=0.4, figsize=(10,7),
)
plt.legend()
plt.show()
```

```python
# ยอดหนี้คงค้างสำหรับบัญชีที่ยังมียอดหนี้คงค้างและไม่ตัดหนี้สูญ
data.plot(kind="scatter", x="latitude", y="longitude",
    s=data['ACTIVE_OS']/1000, label="ACTIVE_OS",
    c="ACTIVE_OS", cmap=plt.get_cmap("jet"),
    colorbar=True, alpha=0.4, figsize=(20,10),
)
plt.legend()
plt.show()
```

```python
from bokeh.io import output_file, output_notebook, show
from bokeh.models import (
  GMapPlot, GMapOptions, ColumnDataSource, Circle, LogColorMapper, BasicTicker, ColorBar,
    DataRange1d, PanTool, WheelZoomTool, BoxSelectTool, Range1d, 
)
from bokeh.models.mappers import ColorMapper, LinearColorMapper
from bokeh.palettes import Viridis5, Plasma256
```

```python
source = ColumnDataSource(
    data=dict(
        lat=data.latitude.tolist(),
        lon=data.longitude.tolist(),
        size=data.ACTIVE_OS_NEW_2.tolist(), # ACTIVE_OS # Normalized (0-1)
        color=data.ACTIVE_OS_NEW.tolist() # ACTIVE_ACCTS
    )
)
```

```python
color_mapper = LinearColorMapper(palette=Plasma256)
circle = Circle(x="lon", y="lat", size="size", fill_color={'field': 'color', 'transform': color_mapper}, fill_alpha=0.5, line_color=None)
plot.add_glyph(source, circle)
```

```python
color_bar = ColorBar(color_mapper=color_mapper, ticker=BasicTicker(),
                     label_standoff=12, border_line_color=None, location=(0,0))
plot.add_layout(color_bar, 'right')
```
