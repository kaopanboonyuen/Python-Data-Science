## a price prediction of used cars posted on Ebay in Germany. 

### The file “car.csv” is modified from the original version, where the target variable is “price”. Details of variables are as below.

1.	ID: car index which is primary key
2.	dateCrawled : when this ad was first crawled, all field-values are taken from this date
3.	name : "name" of the car
4.	seller : private or dealer
5.	offerType
6.	price : the price on the ad to sell the car
7.	abtest
8.	vehicleType
9.	yearOfRegistration : at which year the car was first registered
10.	gearbox
11.	powerPS : power of the car in PS
12.	model
13.	kilometer : how many kilometers the car has driven
14.	monthOfRegistration : at which month the car was first registered
15.	fuelType
16.	brand
17.	notRepairedDamage : if the car has a damage which is not repaired yet
18.	dateCreated : the date for which the ad at ebay was created
19.	nrOfPictures : number of pictures in the ad 
20.	postalCode
21.	lastSeenOnline : when the crawler saw this ad last online
